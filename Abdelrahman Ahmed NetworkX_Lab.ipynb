```python
import networkx as nx
```

```python
#Draw Graph one

G = nx.Graph()
nodes = [0, 1, 2, 3 , 4 , 5 , 6 , 7 ,8]
edges= [(0, 1), (0,3), (0,8), (8,4), (4,3), (3, 2) , (1,7) , (7, 2) , (2,5) , (5,6)]
G.add_node(len(nodes)-1)
G.add_nodes_from(nodes)
G.add_edges_from(edges)
nx.draw(G, with_labels = True)
```

```python
#BFS
list(nx.edge_bfs(nx.Graph(edges), nodes))
```

```python
#BFS other solution

edges = nx.bfs_edges(G, 0)
nodes_bfs = [0] + [v for u, v in edges]
nodes_bfs
```

```python
#DFS
list(nx.dfs_labeled_edges(G , source =0))
```

```python
#DFS other solution

edges = nx.dfs_edges(G, 0)
nodes_bfs = [0] + [v for u, v in edges]
nodes_bfs
```

```python
#Draw Weigthed Graph
G1 = nx.Graph()

G1.add_edge(0, 2, weight=9)
G1.add_edge(0, 1, weight=8)
G1.add_edge(1, 6, weight=9)
G1.add_edge(1, 5, weight=7)
G1.add_edge(1, 3, weight=1)
G1.add_edge(2, 6, weight=4)
G1.add_edge(2, 5, weight=6)
G1.add_edge(3, 7, weight=4)
G1.add_edge(5, 7, weight=2)
G1.add_edge(6, 7, weight=6)
G1.add_edge(6, 4, weight=7)
G1.add_edge(4, 7, weight=7)
nx.draw(G1 , with_labels = True)
```

```python
#Dijkstra

length ,path = nx.single_source_dijkstra(G1, source = 0)
print(length)
path
```

```python
# kruskal
sorted(list(nx.minimum_spanning_edges(G1,algorithm='kruskal' , data = False) ))
```

```python
#Prim
sorted(list(nx.minimum_spanning_edges(G1 ,algorithm='prim', data = False)))
```
